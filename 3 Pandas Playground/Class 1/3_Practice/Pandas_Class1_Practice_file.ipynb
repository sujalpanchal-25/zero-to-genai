## 3 Understanding Series & DataFrame
- **Series**: One-dimensional labeled array, like a column in Excel.
- **DataFrame**: Two-dimensional table, similar to an Excel sheet or SQL table.

### Creating a Series from Python Lists or Dictionaries

````python
a = [10,20,30,40,50]
s = pd.Series(a,name='num')
print(s)
d = {"Sujal":90,"Harsh":20,"Dhruv":80,"Riddhi":70,"kanishka":85}
d = pd.Series(d,name = "Marks")
d
````

### Creating a DataFrame

````python
s = {
    'name': ['Rahul', 'Priya', 'Amit', 'Neha', 'Karan'],
    'subject': ['Math', 'Science', 'English', 'Computer', 'Physics'],
    'marks': [90, 85, 78, 92, 88]
}
pd.DataFrame(s)
````

### Creating DataFrame from NumPy Arrays

````python
a = np.arange(9).reshape(3,3)
df =  pd.DataFrame(a,columns=["A","B","C"])
df
````
