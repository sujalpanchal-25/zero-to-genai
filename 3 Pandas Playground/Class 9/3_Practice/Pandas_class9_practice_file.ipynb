## 1. Applying Custom Functions with `apply()` / `map()`
- `apply()` lets you apply a custom function across rows or columns.
- `map()` is for applying functions element-wise to a Series.
- Both are handy for transformations, calculations, and cleaning data.

```python
import pandas as pd
import numpy as np
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Emma"],
    "Math": [88, 75, 92, 60, 85],
    "Science": [90, 80, 78, 70, 95]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
df['total Marks'] = df[['Math','Science']].apply(lambda row : row.sum(),axis = 1)
```

```python
df
```

```python
col_mean = df[['Math','Science']].apply(lambda col : col.mean())
```

```python
col_mean
```

```python
def grade(s):
  if s>=90:
    return "A"
  elif s>=75:
    return "B"
  if s>=60:
    return "C"
  if s>=55:
    return "D"
  else:
    return "FAIL"
```

```python
df['MathGrade'] = df['Math'].apply(grade)
```

```python
df['ScienceGrade'] = df['Science'].apply(grade)
```

```python
df
```

```python
df['Name'] = df['Name'].map(str.lower)
```

```python
df
```

```python
df['bonus marks'] = df['total Marks'].map(lambda x : x+10)
```
