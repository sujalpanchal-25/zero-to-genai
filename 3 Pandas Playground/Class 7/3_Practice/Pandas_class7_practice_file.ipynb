## 1. Using `concat()` for Stacking Data
- `pd.concat()` combines DataFrames vertically (rows) or horizontally (columns).
- Use `axis=0` for vertical stacking (default), `axis=1` for horizontal.
- `ignore_index=True` can reset the index after concatenation.

```python

# Dataset 1: Product details


# Dataset 2: More products (same columns as df1 → good for vertical stacking)

# Dataset 3: Additional info (different columns → good for horizontal stacking)
```

```python
import pandas as pd

# Dataset 1: Product details
df1 = pd.DataFrame({
    "ProductID": [101, 102, 103],
    "ProductName": ["Pen", "Notebook", "Marker"],
    "Price": [10, 50, 30]
})

# Dataset 2: More products (same columns as df1 - good for vertical concatenation)
df2 = pd.DataFrame({
    "ProductID": [104, 105],
    "ProductName": ["Pencil", "Eraser"],
    "Price": [5, 8]
})

# Dataset 3: Additional information (different columns - good for horizontal concatenation)
df3 = pd.DataFrame({
    "Stock": [100, 60, 80],
    "Rating": [4.5, 4.7, 4.8]
})
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],ignore_index=True)
```

```python
pd.concat([df1,df3],axis=0)
```

```python
pd.concat([df1,df3],axis=1)
```

```python
pd.concat([df1,df3],axis=1,ignore_index=False)
```

## 2. Using `merge()` for SQL-Style Joins
- `merge()` is similar to SQL joins.
- Specify `on` or `left_on`/`right_on` to choose key columns.
- Join types: `inner` (default), `left`, `right`, `outer`.

```python
product = pd.concat([df1,df2],ignore_index=False)
```

```python
product
```

```python
sales = pd.DataFrame({
    "ProductID":[101,102,104,106],
    "UnitsSold":[500,200,150,50]
})
```

```python
pd.merge(product,sales)
```

```python
pd.merge(product,sales,on='ProductID',how='inner')
```

```python
pd.merge(product,sales,on='ProductID',how='left')
```

```python
pd.merge(product,sales,on='ProductID',how='right')
```

```python
pd.merge(product,sales,on='ProductID',how='outer')
```

```python
pd.merge(product,sales,on='ProductID',how='inner')
```
