##Fancy Indexing
you can Pick elements from specific positions using a list/array of indices.

```python
import numpy as np
arr = np.array([10,20,30,40,50,60])
print(arr[[0,2,4]])
```

You can do this fancy indexing in 2D Matrix aswell

```python
arr = np.arange(1,13).reshape(3,4)
print(arr)
print()
print(arr[[0,2,2],[0,1,3]])
```

##Conditional Selection
Find values based on conditions or replace them.

```python
arr = np.array([18,55,45,25,12,8])
print(arr>20)
print(np.where(arr>20))
print(np.where(arr>20,arr,0))
```

##Sorting & Ordering
np.sort() → returns sorted array

np.argsort() → returns indices of sorted order

```python
arr = np.array([25,12,45,8,10,50])
print(np.sort(arr))
```

```python
print(np.argsort(arr))
```

##Concatenation & Stacking

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

np.concat([a,b])
```

Vertical & Horizontal Stacking

```python
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])
print(A)
print(B)
```

```python
print(np.vstack([A,B]))
```

```python
print(np.vstack([B,A]))
```

```python
print(np.hstack([A,B]))
```

```python
print(np.hstack([B,A]))
```
